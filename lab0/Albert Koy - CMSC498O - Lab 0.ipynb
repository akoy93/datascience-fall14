```python
# Render our plots inline
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('UCBAdmissions.csv')
data
```

```python
print data[['Admit', 'Gender']].head()
```

```python
# filter rows by admitted males and admitted females
males_admitted = data[(data.Admit == 'Admitted') & (data.Gender == 'Male')]
females_admitted = data[(data.Admit == 'Admitted') & (data.Gender == 'Female')]
```

```python
males_admitted
```

```python
females_admitted
```

```python
# get number of rows
num_males_admitted = males_admitted.shape[0]
num_females_admitted = females_admitted.shape[0]
```

```python
labels = 'Males', 'Females'
sizes = [num_males_admitted, num_females_admitted]

plt.pie(sizes, labels=labels, autopct='%i%%')
plt.title("Admitted Students by Gender")
```
