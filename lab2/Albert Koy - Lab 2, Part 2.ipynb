```python
import pandas as pd
%pylab inline
```

```python
players = pd.read_csv('players.csv')
countries = pd.read_csv('countries.csv')
events = pd.read_csv('events.csv')
```

```python
# Count the total number of players whose names start with a vowel
players.name.str.extract('^([AEIOU])').count()
```

```python
# Find players from 'USA' whose names start with 'A'.
players[(players.country_id == 'USA') & (players.name.str.contains('^A'))]
```

```python
# Construct a dataframe with two columns: country_name, num_players. Use groupby.
temp_counts = pd.DataFrame({'num_players' : players.groupby('country_id').size()}).reset_index()
player_counts = pd.merge(countries, temp_counts, on='country_id')[['name', 'num_players']].rename(columns={'name' : 'country_name'})
player_counts
```

```python

counts = pd.DataFrame({'count' : players[players.birthdate.str.contains('^1975')].groupby('country_id').size()}).reset_index()
final = pd.merge(countries, counts, how='left', on='country_id')[['name', 'count']]
```
